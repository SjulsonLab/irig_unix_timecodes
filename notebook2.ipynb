```python
import numpy as np
filename = 'data/timestamps_events_absolute (3).csv'
#switch with data/timestamps_events_absolute (3).csv for extra analysis

events = np.genfromtxt(filename, delimiter=',', usecols=3, dtype=str)
times = np.genfromtxt(filename, delimiter=',', usecols=0, dtype=str)

```

```python
frame_times = []
for event, time in zip(events,times):
    if event[:5] == 'frame': 
        frame_times.append(time)

frame_diffs = np.diff(np.array(frame_times, dtype=np.float64))
```

```python
import matplotlib.pyplot as plt

# plt.plot(frame_diffs)
```

```python
lick_times = []
for event, time in zip(events,times):
    if event[:5] == 'Licks': 
        lick_times.append(time)

lick_diffs = np.diff(np.array(lick_times, dtype=np.float64))

count = 0
for diff in lick_diffs:
    if abs(diff - 0.5) > 0.001:
        count += 1

count /= len(lick_diffs)
print(f'Percent off by more than 1ms: {count * 100}%')

plt.hist(lick_diffs, density=True, bins=100, range=(0,1))
```

```python
rising_times = []
for event, time in zip(events,times):
    if event[:5] == 'TimeP': 
        rising_times.append(time)

rising_diffs = np.diff(np.array(rising_times, dtype=np.float64))

count = 0
for diff in rising_diffs:
    if abs(diff - 1) > 0.001:
        count += 1

count /= len(rising_diffs)
print(f'Percent off by more than 1ms: {count * 100}%')

plt.hist(rising_diffs, density=True, bins=50)
```

```python
falling_times = []
for event, time in zip(events,times):
    if event[:5] == 'TimeN': 
        falling_times.append(time)

if falling_times[0] < rising_times[0]:
    falling_times.pop(0)
if falling_times[-1] < rising_times[-1]:
    rising_times.pop(-1)

pulse_lengths = np.array(falling_times, dtype=np.float64) - np.array(rising_times, dtype=np.float64)

plt.hist(pulse_lengths, density=True, bins=100, range=(0,1))
```
